```python
import requests
import re
from datetime import datetime

# Sample URL
url = 'https://shopee.com.my/ANCONA-Powerbank-Original-20000mAh-Pawer-Bank-Original-10000mAh-Power-bank-Murah-Fast-Charging-%E5%85%85%E7%94%B5%E5%AE%9D%E5%BF%AB%E5%85%85-i.264241189.7057509127?sp_atk=691b5244-21a0-43d0-9296-d918a0e4a819&xptdk=691b5244-21a0-43d0-9296-d918a0e4a819'

# Extracting shop_id and item_id from the URL using regex
r = re.search(r'i\.(\d+)\.(\d+)', url)
if r:
    shop_id, item_id = r.group(1), r.group(2)
else:
    print("Invalid URL")
    exit()

# Initializing offset and page_count variables
offset = 0
page_count = 0

# Loop to fetch ratings data from Shopee API
while True:
    # Constructing the ratings API endpoint URL
    ratings_url = f'https://shopee.com.my/api/v2/item/get_ratings?filter=0&flag=1&itemid={item_id}&limit=20&offset={offset}&shopid={shop_id}&type=0'

    # Fetching ratings data from the API and converting the response to JSON format
    data = requests.get(ratings_url).json()

    # Looping through each rating in the fetched data
    for rating in data['data']['ratings']:
        # Printing reviewer name
        print("Reviewer name:", rating['author_username'])

        # Extracting and formatting review date
        review_date = datetime.fromtimestamp(rating['ctime']).strftime('%Y-%m-%d %H:%M:%S')
        print("Review date:", review_date)

        # Printing review content
        print("Review content:", rating['comment'])

        # Printing separator line
        print('-' * 80)

    # Incrementing page_count
    page_count += 1

    # Checking if the maximum page count of 5 is reached
    if page_count == 5:
        break

    # Checking if the number of ratings fetched is less than 20
    if len(data['data']['ratings']) < 20:
        break

    # Incrementing offset for the next page of ratings
    offset += 20
```

```python
import csv

# Create a list of rows to be written to the CSV file
rows = []

# Extract the relevant information from the preceding code
for rating in data['data']['ratings']:
    row = [
        rating['author_username'],
        review_date,
        rating['comment']
    ]
    rows.append(row)

# Open the CSV file in write mode
with open('reviews.csv', 'w', newline='') as csvfile:
    # Create a CSV writer object
    writer = csv.writer(csvfile)

    # Write the header row
    writer.writerow(['Reviewer Name', 'Review Date', 'Review Content'])

    # Write the data rows
    writer.writerows(rows)

# Print a message to indicate that the data has been saved
print("Review data has been saved to the CSV file 'reviews.csv'")
```
